### Initialization of Tensors

```python
x = tf.constant(4.0, shape=(1,1),dtype=tf.float32)
y = tf.constant([[1,2,3],[4,5,6]])
z = tf.ones((3,3))
a = tf.zeros((2,3))
b = tf.eye(3) # I for the identity matrix (eye)
c = tf.random.normal((3,3), mean=0 , stddev=11)
d = tf.random.uniform((1,3), minval=0 , maxval=1)
e = tf.range(9)
f = tf.range(start=1,limit=10,delta=2)
g = tf.cast(g,dtype=tf.float64)

# tf.float(16,32,64) ; tf.int(8,16,32,64) ; tf.bool
```

### Mathematical Operations

```python
x = tf.constant([1,2,3])
y = tf.constant([9,8,7])
```

```python
#Addition
z = tf.add(x,y)
z = x + y
```

```python
print(z)
```

```python
#Subtraction
z = tf.subtract(x,y)
z = x - y
```

```python
print(z)
```

```python
# Division
z = tf.divide(x,y)
z = x / y
```

```python
print(z)
```

```python
# Multiplication
z = tf.multiply(x,y)
z = x * y
```

```python
print(z)
```

```python
# Dot product
z = tf.tensordot(x,y,axes=1)
z = tf.reduce_sum(x*y,axis=0)
```

```python
print(z)
```

```python
# Power product
z = x ** 5
```

```python
print(z)
```

```python
# Matrix multiplication
x = tf.random.normal((2,3))
y = tf.random.normal((3,4))
z = tf.matmul(x,y)
z = x @ y
```

```python
print(z)
```

### Indexing

```python
x = tf.constant([0,1,1,2,3,1,2,3])
```

```python
print(x[:])
```

```python
print(x[1:])
```

```python
print(x[1:3])
```

```python
print(x[::2])
```

```python
print(x[::-1])
```

```python
indices = tf.constant([0,3])
x_ind = tf.gather(x,indices) #vector
print(x_ind)
```

```python
# Matrix

x = tf.constant([[1,2],[3,4],[5,6]])
```

```python
print(x[0,:1])
```

```python
print(x[0:2,:])
```
